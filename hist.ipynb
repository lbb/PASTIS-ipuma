```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
cs = np.loadtxt("0slenc.txt", dtype=np.int64)
rs = np.loadtxt("0slenr.txt", dtype=np.int64)

cs1 = np.loadtxt("1slenc.txt", dtype=np.int64)
rs1 = np.loadtxt("1slenr.txt", dtype=np.int64)
```

```python
crs0 = np.concatenate((cs, rs))
crs1 = np.concatenate((cs1, rs1))
bins = np.linspace(0, max(max(crs0), max(crs1)), num=200)



plt.hist(crs0, bins, alpha=0.5, label='crs0')
plt.hist(crs1, bins, alpha=0.5, label='crs1')
plt.legend(loc='upper right')
plt.show()
```
